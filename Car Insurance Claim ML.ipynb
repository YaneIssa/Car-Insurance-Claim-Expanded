```python
# 1. Regression: Predicting CREDIT_SCORE 
# Import required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pandas as pd

# Load dataset
df = pd.read_csv("Car_Insurance_Claim.csv")

# Select relevant columns for Regression
df = df[['CREDIT_SCORE', 'AGE', 'INCOME', 'ANNUAL_MILEAGE', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS']]

# Convert categorical columns to numerical
df['AGE'] = df['AGE'].astype('category').cat.codes
df['INCOME'] = df['INCOME'].astype('category').cat.codes

# Check missing values before cleaning
print("Missing values before cleaning:\n", df.isnull().sum())

# Create a new cleaned DataFrame
df_clean = df.copy()  # Copy to avoid modifying the original data

# Fill missing values with median
df_clean['CREDIT_SCORE'] = df_clean['CREDIT_SCORE'].fillna(df_clean['CREDIT_SCORE'].median())
df_clean['ANNUAL_MILEAGE'] = df_clean['ANNUAL_MILEAGE'].fillna(df_clean['ANNUAL_MILEAGE'].median())

# Check missing values after cleaning
print("Missing values after cleaning:\n", df_clean.isnull().sum())

# Define features (X) and target (Y)
X = df_clean.drop(columns=['CREDIT_SCORE'])
Y = df_clean['CREDIT_SCORE']

# Split dataset into training (80%) and testing (20%) sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predict credit scores
Y_pred = model.predict(X_test)

# Calculate accuracy using R² Score
print("R² Score:", r2_score(Y_test, Y_pred))
```

```python
# 2.Classification: Predicting OUTCOME (Insurance Claim)
# Import required libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Load dataset
df = pd.read_csv("Car_Insurance_Claim.csv")

# Select relevant columns for Classification
df = df[['OUTCOME', 'GENDER', 'DRIVING_EXPERIENCE', 'EDUCATION', 'INCOME', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS']]

# Drop missing values
df = df.dropna()

# Convert categorical columns to numerical
for col in ['GENDER', 'DRIVING_EXPERIENCE', 'EDUCATION', 'INCOME']:
    df[col] = df[col].astype('category').cat.codes

# Define features (X) and target (Y)
X = df.drop(columns=['OUTCOME'])
Y = df['OUTCOME']

# Split dataset into training (80%) and testing (20%) sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train a Random Forest Classification model
#model = RandomForestClassifier(n_estimators=100, random_state=42)
#model.fit(X_train, Y_train)

# Define number of neighbors before using it
k = 5  # You can change this value (e.g., k=3, k=7)

# Train a K Neighboor Classification model
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, Y_train)

# Predict insurance claims
Y_pred = model.predict(X_test)

# Calculate accuracy
print("Accuracy Score:", accuracy_score(Y_test, Y_pred))
```

📌 Think of Cluster 2 as "Safe & Responsible Drivers" – financially stable and cautious.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("Car_Insurance_Claim.csv")

# Select relevant columns for clustering
features = ['CREDIT_SCORE', 'ANNUAL_MILEAGE', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS']
df_selected = df[features].dropna()  # Remove missing values

# Scale the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)

# Apply PCA (Reduce to 2 components for visualization)
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_pca)

# Convert PCA output to DataFrame
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = clusters

# Scatter Plot: PCA Components
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_pca['PC1'], y=df_pca['PC2'], hue=df_pca['Cluster'], palette='viridis', alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Clusters Based on PCA Transformed Features")
plt.legend(title="Cluster")
plt.show()
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=features)

# Display PCA Loadings
print("PCA Loadings:")
print(loadings)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Select relevant features
features = ['CREDIT_SCORE', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS', 'INCOME']

# Handle categorical variables (e.g., 'INCOME' if it's an object type)
if df['INCOME'].dtype == 'object':  
    le = LabelEncoder()
    df['INCOME'] = le.fit_transform(df['INCOME'])

# Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Replace NaNs with column mean
df[features] = imputer.fit_transform(df[features])

# Define feature variables (X) and target variable (y)
X = df[features]
y = df['OUTCOME']

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(f"Model Accuracy: {accuracy_score(y_test, y_pred):.2%}")

```
